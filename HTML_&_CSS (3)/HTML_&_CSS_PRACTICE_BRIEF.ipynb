# TASK 1 & 2

```python
url="https://en.wikipedia.org/wiki/Leonardo_DiCaprio_filmography"
html_content=requests.get(url).text
soup=bs(html_content,"lxml")
print(soup.prettify())
```

```python
#defined table at first...
table = soup.find('table')

new_table=[]
for row in table.find_all('tr')[4:]:  #tr = table rows...
    columns = row.find_all('td')
    new_table.append([column.text for column in columns[0:]])
new_table
```

```python
df=pd.DataFrame(new_table, columns=["Title","Role","Notes","Ref"])
df[:5]
```

```python
#replacing the "\n" values to "empty spaces" in the columns...
df["Title","Role","Notes","Ref"]=df["Title"].str.replace("\n","")
df["Role"]=df["Role"].str.replace("\n","")
df["Notes"]=df["Notes"].str.replace("\n","")
df["Ref"]=df["Ref"].str.replace("\n","")
```
