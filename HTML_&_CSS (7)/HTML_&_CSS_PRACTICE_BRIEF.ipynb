# TASK 2

```python
#phrase2="I prefer the morning flight through Denmark"

doc = nlp("I prefer the morning flight through Denmark")                   
print(doc)  

for chunk in doc.noun_chunks:
   print('"Text"=',chunk.text,
         '\t"Root text"=',chunk.root.text,
         '\t"Root dep"=',chunk.root.dep_,
          '\t"Root head text"=',chunk.root.head.text)
```

>Text:- The original noun chunk text.  
Root text:- The original text of the word connecting the noun chunk to the rest of the parse.   
Root dep:- Dependency relation connecting the root to its head. nsubj: Nominal subject. The details of such terms are given in learning consolidation.  
Root head text:- The text of the root token’s head.

##### Removing stop-words:

```python
spacy_stopwords = spacy.lang.en.stop_words.STOP_WORDS

from spacy.lang.en.stop_words import STOP_WORDS            

filtered_file=[] 

doc = nlp(text) 

for word in doc:                                 
    if word.is_stop==False:                    
        filtered_file.append(word)                  
print("\n \nFiltered File:",filtered_file)         
```

##### Lematization:

```python
for word in filtered_file:                
    print(word.text,"-----",word.lemma_)
```

##### Parts Of Speech (POS):

```python
file = open("Raw_data_for_analysis.txt")

contents = file.read()                                 
text = str(contents)                                              
doc = nlp(text)

for word in doc:          
    print(word.text,
          '\t',word.pos_,
          '\t',word.tag_)
```

##### Named Entity Recognition (NER):

```python
file = open("Raw_data_for_analysis.txt")

contents = file.read()                                 
text = str(contents)                                              
doc = nlp(text)

for ent in doc.ents:
    print(ent.text, '\tstart=',
          ent.start_char, '\tend=',
          ent.end_char, '\tLabel=',
          ent.label_)
```

* Labels of :

1. Michael ----- PERSON
2. School ----- ORG
3. 1997 ----- DATE
